### Counting!

What function can you use to find the number of rows and/or columns in a DataFrame?

**Answer**: len()

Find the number of columns in ```host_cities``` using your answer above. Print the number of columns.

````python
print(len(host_cities.columns))
````

Find the number of rows in ```host_cities```. Print the answer.

````python
print(len(host_cities))
````

### VIP Access

Replace the lines below with your answers to make the sentences correct.

```.index``` will show you the different row labels (the bolded numbers all the way to the left) of a DataFrame and ```.values``` will show you all the data. While they may look like functions, they are not, as they do not have parentheses.

The function ```.info()``` does something similar to ```.values``` but it provides more detail on the type of each column rather than the values of the DataFrame.

### Selecting Rows

What method can you call on your DataFrame that uses the row's index to access that row?

**Answer**: ```.iloc[]```

What method can you call on your DataFrame that uses the row's label to access that row?

**Answer**: ```.loc[]```

Unfortunately, in this DataFrame ```host_cities```, the row labels and row indexes are the same. (So many tables have the same values for the row index and label. I honestly tried finding one where the label differed from the index but . . . as you can see, I failed. Oh well.) So . . . we're basically going to pretend that they're different. Imagine that the leftmost, bolded column (the one that labels the rows) has ```"city_name, country_name"``` values (ex: the label for the first row would be "Athens, Greece") instead of numbers. <br> <br> Access the row whose city is Vancouver using ```.loc[]```. (Maybe don't run the cell. Because it'll error as we are playing pretend.)

````python
host_cities.loc["Vancouver, Canada"]
````

Access the rows starting from Squaw Valley to Muncih using ```.loc[]```. (Again, maybe don't run the cell.)

````python
host_cities.loc["Squaw Valley":"Munich"]
````

Now, access the 27th row using ```.iloc[]```. (You can run this cell)

````python
host_cities.iloc[26]
````

Access all the rows whose index is between 10 and 20 (inclusive) in ```host_cities``` using ```.iloc[]```.

````python
host_cities.iloc[10:21]
````

### Selecting Columns

Replace the lines below with your answers to make the sentence correct.

To select a column inside a DataFrame, you want to use ```bracket``` notation like you did back when you were using dictionaries, and pass in the ```name``` of the column as a String.

Access the column in ```host_cities``` that holds the city's name.

````python
host_cities["City"]
````

Now, access the columns that have the city's name, the country's name, and the year.

````python
host_cities[["City","Country","Year"]]
````
