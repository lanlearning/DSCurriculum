# Terminology about Data: 

### Numerical Data
Data that is made up of numbers!

**Example**:
- height, weight
- number of sales made
- time it takes to run a mile

### Categorical Data
Data that is split into categories!

**Example**:
- grade level of students
- birth month (still related to time)
- car type based on number of wheels

### Distribution
Distirbutions are used to see how much data the data varies. Essentially, gives you an idea about the data you have.

# Histogram

### Used to plot the distribution of a numerical variables.
*This plot looks like a mountain and has areas for each numerical interval range.*

<br>
<img src='../images/histp.png' width="900"/>
<br>

**Here, the highest point of the mountain represents the most often occuring point.**

````python
astro.head()
````

````python
# making the histogram using sns.distplot(a= ) where a is some array of values
 
# Let's plot the distribution of the number of space walks in our astro dataframe

# Think about why we don't need to group our DataFrame for the histogram?

sns.distplot(astro['Space Walks'], color='saddlebrown');
````

### How to make meaning of the histogram:

From the histogram, the mountaintop is at 0, which means the most number of space walks for a certain astronaut was 0. There are a few astronauts with 2, 4, 6, and 6+ space walks, but the number of those astronauts is less than the number of astronauts with 0 space walks. That's why the mountain starts to get smaller near bigger numbers for space walks. 

The situation where the mountain gets smaller is know as **skew** where we say skew left if the mountain gets smaller towards the left, and skew right if smaller at the right. 

The plot above is **skew right**

````python
astro['Space Flight (hr)'].value_counts()
````

````python
# another distplot (distribution plot)

sns.distplot(astro['Space Flight (hr)'], hist=False, color='black')
````

### No y-axis:

The y-ais gives you information about the density of values at a certain point. For example, the density for the mountaintop is the highest because the most points are of that x-value.

With Seaborn, your DataFrame **needs to prepared a certain way** for it to work with the heatmap function provided by sns.

**I'm going to walk through something I did before**:

````python
players = pd.read_csv('20.csv')
print(len(players))
players.head()
````

````python
# selecting the all star players from the players DataFrame: 

allstar = ['LeBron', 'Kawhi Leonard', 'Giannis', 'Pascal', 'Chris Paul',
           'Doncic', 'Harden', 'Anthony Davis', 'Lillard', 'Booker', 'Jokic',
           'Ben Simmons', 'Sabonis', 'Tatum', 'Westbrook', 'Embiid', 'Kemba Walker',
           'Trae Young', 'Adebayo', 'Ingram', 'Donovan Mitchell', 'Jimmy Butler',
           'Gobert', 'Lowry', 'Middleton']

def all_star(i):
    return any([j in i for j in allstar])

allsta = players[players['Player'].apply(all_star)]
print(len(allsta))
allsta.head()
````

````python
allsta = allsta[['Player', 'MP', 'PTS', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', 'TRB', 'AST',
                 'STL', 'BLK', 'TOV']]
allsta.head()
````

````python
len(allsta.columns)
````

````python
name = 'Bam Adebayo\adebaba01'

name.split('\\')    #[0].split(' ')[1]
````

````python
def name(i):
    return i.split('\\')[0].split(' ')[1]

allsta['Player'] = allsta['Player'].apply(name)
allsta.head()
````

````python
# for heatmap to work, you need to have your y-labels be your index

# i want my y-labels to be the players' names, so I need to make that the index
allsta.index = allsta['Player']
allsta.head()
````

````python
allsta = allsta.drop('Player', axis=1)
````

````python
allsta.head()
````

````python
allsta = allsta.fillna(0)
````

````python
sns.set(rc={'figure.figsize':(11.7,7)})
sns.heatmap(allsta, cmap='Purples')
prop = fm.FontProperties(fname='Liberator-Light.ttf')
plt.xticks(fontproperties=prop, size=16);
plt.yticks(fontproperties=prop, size=16);
plt.gca().get_yticklabels()[10].set_color("goldenrod");
plt.tight_layout()
plt.savefig('allstars');
````

# The plot above is bad

Because the values for the different columns vary. For example, 3P% will always be between 0 and 1, wheras PTS may be a number as big as 30. This will give us problems when coloring the squares so we will **normalize** the columns! 

**Normalizing** is a technique to clean your data and be able to get proportions from them. 

### Take each value and divide by the sum of that column:

````python
1 + 2 + 3 + 4
````

````python
4/10
````

````python
col_normalize = ['MP', 'PTS', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', 'TRB',
       'AST', 'STL', 'BLK', 'TOV']

for i in col_normalize:
    col_sum = sum(allsta[i])
    allsta[i] = allsta[i]/col_sum
    
allsta.head()
````

````python
sns.set(rc={'figure.figsize':(11.7,7)})
cmap = sns.cm.rocket_r
sns.heatmap(allsta, cmap='Purples')
prop = fm.FontProperties(fname='Liberator-Light.ttf')
plt.xticks(fontproperties=prop, size=16);
plt.yticks(fontproperties=prop, size=16);
plt.gca().get_yticklabels()[10].set_color("goldenrod");
plt.tight_layout()
plt.savefig('allstars');
````

````python
# For cleaning it up and organizing it: sort the players by ppg and then plot: 

sns.set(rc={'figure.figsize':(11.7,7)})
cmap = sns.cm.rocket_r
sns.heatmap(allsta.sort_values('PTS', ascending=False), cmap='Purples')
prop = fm.FontProperties(fname='Liberator-Light.ttf')
plt.xticks(fontproperties=prop, size=16);
plt.yticks(fontproperties=prop, size=16);
plt.gca().get_yticklabels()[10].set_color("goldenrod");
plt.tight_layout()
plt.savefig('allstars');
````
